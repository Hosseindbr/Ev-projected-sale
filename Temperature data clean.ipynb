```python
# Import dependencies
import pandas as pd
import numpy as np
import json 
import requests
import timeit
from datetime import datetime
import matplotlib.pyplot as plt
from meteostat import Stations, Daily
```

```python
#pip install meteostat
# Import Meteostat library
from meteostat import Point

start_date = datetime(2015, 1, 1)
end_date = datetime(2021, 12, 31)
# Create Point for New York City, NY
New_York_City = Point(40.71427, -74.00597)

# Get daily data
data = Daily(New_York_City, start_date, end_date)
data = data.fetch()

# Show dataframe
data=data.reset_index()
data
```

```python
data['time'] = pd.to_datetime(data['time'])
data.head()
```

```python
start_date = '2015-01-01'
end_date = '2021-12-31'
mask = (data['time'] >= start_date) & (data['time'] <= end_date)
NY_state_annual_temperature= data.loc[mask]
NY_state_annual_temperature
```

```python
#adding the column year
NY_state_annual_temperature['Year'] = NY_state_annual_temperature['time'].dt.strftime("%Y")
NY_state_annual_temperature.head()
```

```python
#only keeeping the columns we need as the time, tavg and year
NY_state_annual_temperature=NY_state_annual_temperature[['time','tavg','Year']]
NY_state_annual_temperature
```
