```python
# Import dependencies
import pandas as pd
import numpy as np
import json 
import requests
import timeit
from datetime import datetime
```

# We need to read all of our dataset

```python
#Registrations data set
NYS_registrations_df=pd.read_csv('Resources/NYS_EV_Registrations.csv')
NYS_registrations_df = NYS_registrations_df.rename(columns={"Reg Valid Date": "Reg_Valid_Date", "Model Year": "Model_Year", "Registration year": "Registration_year"})
NYS_registrations_df.drop('Unnamed: 0', axis=1, inplace=True)
NYS_registrations_df
```

```python
NYS_registrations_df.dtypes
```

```python
# Demographic charging stations
Charger_df=pd.read_csv('Resources/Demographic_Database_EV_Stations.csv')
Charger_df = Charger_df.rename(columns={"Station Name": "Station_Name"})
Charger_df['County'] = Charger_df['County'].str.upper()
Charger_df.drop('Unnamed: 0', axis=1, inplace=True)
Charger_df
```

```python
Charger_df.dtypes
```

```python
#Income and Population dataset
Income_and_Population_df=pd.read_csv('Resources/Income_and_Population_per_Years_2015_2021.csv')
Income_and_Population_df = Income_and_Population_df.rename(columns={"Year": "Year_"})

Income_and_Population_df['County'] = Income_and_Population_df['County'].str.upper()
Income_and_Population_df.drop('Unnamed: 0', axis=1, inplace=True)

Income_and_Population_df.head(5)
```

```python
Income_and_Population_df.dtypes
```

```python
#Energy datasets
NYS_Annual_Energy_Generation_and_Sales=pd.read_csv("Resources/NYS_Annual_Energy_Generation_and_Sales.csv")
NYS_Annual_Energy_Generation_and_Sales = NYS_Annual_Energy_Generation_and_Sales.rename(columns={"YEAR": "YEAR_"})
NYS_Annual_Energy_Generation_and_Sales.drop('Unnamed: 0', axis=1, inplace=True)
NYS_Annual_Energy_Generation_and_Sales.head(5)
```

```python
NYS_Annual_Energy_Generation_and_Sales.dtypes
```

```python
#temperature daily to match registration table
NY_state_annual_temperature_2015_2021=pd.read_csv('Resources/Temperature_for_New_York_City_per_Years_2015_2021.csv') 
NY_state_annual_temperature_2015_2021.drop('Unnamed: 0', axis=1, inplace=True)
NY_state_annual_temperature_2015_2021
```

```python
NY_state_annual_temperature_2015_2021.dtypes
```

```python
#Temperature monthly
NY_state_annual_temperature=pd.read_csv('Resources/NY-state-annual-temperature.csv') 
NY_state_annual_temperature['Date'] = pd.to_datetime(NY_state_annual_temperature['Date'])
NY_state_annual_temperature['Year_'] = NY_state_annual_temperature['Date'].dt.strftime("%Y")
drop_cols =['Unnamed: 0']
NY_state_annual_temperature.drop(drop_cols, axis=1, inplace=True)
NY_state_annual_temperature.head(5)
```

```python
NY_state_annual_temperature.dtypes
```

```python
NY_state_annual_temperature['Year_']= NY_state_annual_temperature['Year_'].astype('int64')
NY_state_annual_temperature.dtypes
```

```python
NY_state_annual_temperature
```

```python
# Saving all the datasets first
NY_state_annual_temperature.to_csv('Resources/NY-state-annual-temperature.csv')
NYS_Annual_Energy_Generation_and_Sales.to_csv("Resources/NYS_Annual_Energy_Generation_and_Sales.csv")
Income_and_Population_df.to_csv('Resources/Income_and_Population_per_Years_2015_2021.csv')
Charger_df.to_csv('Resources/Demographic_Database_EV_Stations.csv')
NYS_registrations_df.to_csv('Resources/NYS_EV_Registrations.csv')
NY_state_annual_temperature_2015_2021.to_csv('Resources/Temperature_for_New_York_City_per_Years_2015_2021.csv')
```

# Importing the merge done in SQL

```python
daily_reg_and_temperature_df=pd.read_sql_table('daily_reg_and_temperature', con=engine)
daily_reg_and_temperature_df
```

```python
daily_reg_and_temperature_df=daily_reg_and_temperature_df.dropna()
daily_reg_and_temperature_df['Model_Year']= daily_reg_and_temperature_df['Model_Year'].astype('Int64')
daily_reg_and_temperature_df['Zip']= daily_reg_and_temperature_df['Zip'].astype('Int64')
daily_reg_and_temperature_df['Registration_year']= daily_reg_and_temperature_df['Registration_year'].astype('Int64')
daily_reg_and_temperature_df
```

```python
daily_reg_and_temperature_df.to_sql('daily_reg_and_temperature', con=engine, index=False, if_exists='replace')
daily_reg_and_temperature_df.to_csv('Resources/daily_reg_and_temperature.csv')
```

```python
energy_sales_and_temperature_df=pd.read_sql_table('energy_sales_and_temperature', con=engine)
energy_sales_and_temperature_df
```

```python
ev_stations_income_pop_df = pd.merge(Charger_df, Income_and_Population_df,  how='left', left_on='County', right_on ='County')
ev_stations_income_pop_df
```

```python
ev_stations_income_pop_df.dtypes
```

```python
ev_stations_income_pop_df['Year_']= ev_stations_income_pop_df['Year_'].astype('Int64')
ev_stations_income_pop_df['Income_US']= ev_stations_income_pop_df['Income_US'].astype('Int64')
ev_stations_income_pop_df['Income_NYS']= ev_stations_income_pop_df['Income_NYS'].astype('Int64')
ev_stations_income_pop_df['Population']= ev_stations_income_pop_df['Population'].astype('Int64')
ev_stations_income_pop_df
```

```python
ev_stations_income_pop_df=ev_stations_income_pop_df.dropna()
ev_stations_income_pop_df
```

```python
ev_stations_income_pop_df.to_sql('ev_stations_income_pop', con=engine, index=False, if_exists='replace')
ev_stations_income_pop_df.to_csv('Resources/ev_stations_income_pop.csv')
```

```python
energy_sales_and_stations_by_income_pop_df = pd.merge(NYS_Annual_Energy_Generation_and_Sales, ev_stations_income_pop_df,  how='left', left_on=['YEAR_','STATE'], right_on =['Year_','State'])
energy_sales_and_stations_by_income_pop_df['City']=energy_sales_and_stations_by_income_pop_df['City'].str.upper()
energy_sales_and_stations_by_income_pop_df
```

```python
drop_cols=['State','Year_']
energy_sales_and_stations_by_income_pop_df.drop(drop_cols, axis=1, inplace=True)
energy_sales_and_stations_by_income_pop_df
```

```python
energy_sales_and_stations_by_income_pop_df.to_sql('energy_sales_and_stations_by_income_pop', con=engine, index=False, if_exists='replace')
energy_sales_and_stations_by_income_pop_df.to_csv('Resources/energy_sales_and_stations_by_income_pop.csv')
```
